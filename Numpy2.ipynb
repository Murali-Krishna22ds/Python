```python
data=np.random.randint(1,10,(5,5))
data
```

```python
d_matrix=[]
for i in data:
    for j in data:
        distance=np.sqrt((((i-j)**2)).sum())
        d_matrix.append(int(distance))
```

```python
np.array(d_matrix).reshape(5,5)
```
