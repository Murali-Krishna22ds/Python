# Whenever there is a categorical column like gender or pass/fail , state/country codes,then we use mode for the null value treatment

## if there are no outliers in the data and it is continuous like sales,profit,units, then we can use mean,median to fill missing values.

#### whenever there is a time series data, data is available in date format then we can use "Forwardfill" ,"Backwardfill", to fill missing values. 
##### Note: sometimes we can split the data and perform both forward fill for weekdays and backward fill for weeekends and try to append data.

```python
### Null value Treatment
 ### check the null values in the dataset
 ### Take the percentage of null values
 ###Confirm the null values
 ### beyond the threshold value we can drop the column
 ### below the threshold value we can keep the column
 ### if it is a numerical column we can fill the null values with mean, median
 ### if it is a categorical column we can fill the null values with mode
 ### if it is a time series data we can fill the null values with forward fill and backward fill
```
