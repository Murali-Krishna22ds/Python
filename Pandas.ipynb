```python
import pandas as pd
import numpy as np

data = {
    'Month': ['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales': [1000, 1500, 1200, 1800, 2000, 1700],
    'Profit': [200, 300, 250, 400, 500, 350],
    'Expenses': [800, 1200, 950, 1400, 1500, 1350]
}

df=pd.DataFrame(data) 
df.set_index('Month', inplace=True)  
df
```

```python
df.columns
```

```python
df[['Sales','Expenses']].plot(title="Line Plot - Sales Trend")
```

```python
df[['Sales','Expenses']].plot(title="Multiple Line Plot")
```

```python
df[['Sales','Expenses']].plot(kind='bar', title="Bar Plot - Sales")
```

```python
df[['Sales','Expenses']].plot(kind='bar', stacked=True, title="Stacked Bar Plot")
```

```python
df[['Sales','Expenses']].plot(kind='barh', title="Horizontal Bar Plot")
```

```python
df[['Expenses','Profit']].plot(kind='hist', bins=5, title="Histogram - Expenses")
```

```python
df[['Sales','Profit']].plot(kind='area', title="Area Plot")
```

```python
df['Sales'].plot(kind='pie', autopct='%1.1f%%', title="Pie Chart - Sales")
```
